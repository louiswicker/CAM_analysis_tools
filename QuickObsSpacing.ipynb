```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

import xarray as xr

import pyproj

import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

import numpy as np

# setup your projections
proj_daymet = "+proj=lcc +lat_0=42.5 +lon_0=-100 +lat_1=25 +lat_2=60 +x_0=0 +y_0=0 +datum=WGS84 +units=m +no_defs" #my custom CRS

import math
def distance(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)
```

```python
file_2020 = '/work2/wof/realtime/2020/20200504/enkfdir/gsi21/vrobs.nc'
file_2021 = '/work2/wof/realtime/2019/20190520/enkfdir/gsi21/vrobs.nc'
```

```python
def mysub(file):
    ds = xr.open_dataset(file)

    x, y = pyproj.Proj(proj_daymet)(ds.lon, ds.lat)
    
    dist = []
    
    for i in range(1, len(x)):
    # you can use print or return, depending on your needs
        dist.append(distance((x[i-1],y[i-1]), (x[i],y[i]) ))
        
    return np.asarray(dist)
```

```python
d2021 = mysub(file_2021)
d2020 = mysub(file_2020)
```

```python
print( d2021.min(), d2020.min() )
```
