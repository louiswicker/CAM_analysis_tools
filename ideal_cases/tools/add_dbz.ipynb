```python
def add_dbz(path, file_pattern=None):

    if file_pattern == None:
        # see if the path has the filename on the end....
        if os.path.basename(path)[:-3] != ".nc":
            path = os.path.join(path, _default_file)
            print(f'-'*120,'\n')
            print(" Added default filename to path input:  %s" % path)

        print(f'-'*120,'\n')
        print(" Reading:  %s \n" % path)
        try:
                ds = xr.load_dataset(path, decode_times=False)
        except:
                print("Cannot find the file in %s, exiting"  % path)
                sys.exit(-1)
                
    dbz_filename = os.path.join(os.path.dirname(path), 'dbz.npz')
    
    

    if os.path.exists(dbz_filename):
        print("\n Reading external DBZ file: %s\n" % dbz_filename)
        with open(os.path.join(os.path.dirname(path), 'dbz.npz'), 'rb') as f:
            dbz = np.load(f)
            
    new = xr.DataArray( dbz, dims = ['nt', 'nz', 'ny', 'nx'] )
    
    ds['dbz'] = new
                       
    ds.to_netcdf(path, mode='w')
    
    print("\n Wrote file:  %s\n" % path)
            
            
```

```python
_default_file = "output_quad.nc"

dirs    = {
           "solo": "/work/wicker/Odin_scr/solo",
           "wrf": "/work/wicker/WRF/WRF/test/em_quarter_ss",
           "cm1": "/work/wicker/Odin_scr/cm1r20.3/run",
           "mpas": "/scratch/wicker/MPAS/ideal/squall",
          }

allcape  = ("C2000", "C3500")
allshear = ( "06", "12", "18")

run = {"mpas": "squall_3km_ic"}

for key in run:
    
    for shear in allshear:
        for cape in allcape:

            label = "%s_%s" % (cape, shear)

            file = str(os.path.join(dirs[key], "%s_%s" % (run[key], label)))
    
            add_dbz(file)
        
        
```
