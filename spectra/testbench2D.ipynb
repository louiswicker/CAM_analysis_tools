```python
import numpy as np
from matplotlib import colors, ticker
import matplotlib.pyplot as plt
import xarray as xr
import os
import sys

from py_spectra import *

file = '/work/wicker/climate_runs/cm1r20.3/run/bubble/3km/QV16/cm1out.nc'

cm1 = xr.load_dataset(file, decode_times=False)
```

```python
# Local import 

fig, ax = plt.subplots(1,2, constrained_layout=True,figsize=(14,7))
ylim    = [0.001,1.0e6]
PSonly  = True

ax[0].pcolormesh(cm1['w'][20,20], shading='auto', vmax=10, vmin=-10, cmap='viridis')

plot_spectra(cm1['w'][20,20], ax = ax[1], axis=ax[1], func=get_spectra2D_DCT, legend='', PScolor='r', ylim=ylim, LinsborgSlope=False, PSonly=PSonly)       
```
